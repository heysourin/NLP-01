```python
from gensim.models import Word2Vec, KeyedVectors
```

```python
import gensim.downloader as api

model = api.load('word2vec-google-news-300')
```

```python
vec_king = model['king']
```

```python
# Vector of the word 'king' is 300 dimensions.
vec_king
```

```python
# Similar words to man
model.most_similar('man')
```

```python
model.most_similar('cricket') # Finding closest words using cosine similarity
```

```python
model.similarity('man', 'king')
```

```python
model.similarity('morning', 'death')
```

```python
# Finds the odd words out: I mean which is not related to the rest.
model.doesnt_match(['Java', 'PHP', 'JavaScript'])
```

```python
# Finds the odd words out: I mean which is not related to the rest.
model.doesnt_match(['Java', 'PHP', 'human'])
```

```python
vec = model['king'] - model['man'] - model['woman']
model.most_similar([vec])
```

```python
vec = model['INR'] - model['India'] - model['England']
model.most_similar([vec])
```
